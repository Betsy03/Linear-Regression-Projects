```python
scalar=StandardScaler()
scalar_data=scalar.fit_transform(employee)
employee1=pd.DataFrame(scalar_data,columns=['salary_hike','churn_rate'])
```

```python
employee1.describe()
```

```python
sns.relplot(x='salary_hike',y='churn_rate',data=employee1,kind='scatter')
```

```python
#Findings
#There is 10 rows and 2 columns.
#No missing Values.
#Churn rate and Salary hike is -ve linear.
#Applied Standart scalar method to scale the values.
```

```python
#Dividing the data into train and test data
x=employee1.salary_hike
y=employee1.churn_rate
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
 X_train
```

```python
X_train=X_train.values.reshape(-1,1)
X_test=X_test.values.reshape(-1,1)
Y_train=Y_train.values.reshape(-1,1)
Y_test=Y_test.values.reshape(-1,1)
```

```python
#Linear Regression
reg=LinearRegression()
reg=reg.fit(X_train,Y_train)
```

```python
Y_pred=reg.predict(X_test)
```

```python
r2_score=reg.score(X_test,Y_test)
```

```python
print(r2_score)
```

```python
plt.plot(employee1.salary_hike,employee1.churn_rate,color="green",label='Regression')
plt.scatter(X_test,Y_pred,marker="+",color="red",label='Scatterplot')
plt.legend()
```
